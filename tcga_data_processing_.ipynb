```python
import pandas as pd
import gzip
import os

maf_list = []

for root, dirs, files in os.walk("."):
    for file in files:
        if file.endswith(".maf.gz"):
            path = os.path.join(root, file)

            try:
                with gzip.open(path, "rt") as f:
                    df = pd.read_csv(f, sep="\t", comment="#", low_memory=False)
                    maf_list.append(df)
            except Exception as e:
                print("Skipped:", path)

print("Files loaded:", len(maf_list))

mutation_df = pd.concat(maf_list, ignore_index=True)

print("Mutation dataframe shape:", mutation_df.shape)
```

```python
mutation_df = mutation_df[[
    "Tumor_Sample_Barcode",
    "Hugo_Symbol",
    "Variant_Classification"
]]

mutation_df.head()
```

```python
mutation_df["Patient_ID"] = mutation_df["Tumor_Sample_Barcode"].str[:12]

mutation_df.head()
```

```python
top_genes = mutation_df["Hugo_Symbol"].value_counts().head(50)

print(top_genes)
```

```python
selected_genes = [
    "IDH1",
    "TP53",
    "ATRX",
    "EGFR",
    "PTEN",
    "NF1"
]

mutation_df = mutation_df[
    mutation_df["Hugo_Symbol"].isin(selected_genes)
]
```

```python
mutation_matrix = pd.crosstab(
    mutation_df["Patient_ID"],
    mutation_df["Hugo_Symbol"]
)

mutation_matrix = (mutation_matrix > 0).astype(int)

mutation_matrix.reset_index(inplace=True)

mutation_matrix.head()
```

```python
print(mutation_matrix.shape)
```

```python
import numpy as np

gbm = pd.read_csv("/content/drive/MyDrive/clinical.project-tcga-gbm.2026-03-03/clinical.tsv", sep="\t")

# Replace missing indicators
gbm = gbm.replace("'--", np.nan)
gbm = gbm.replace("--", np.nan)

gbm_clean = gbm[[
    "cases.submitter_id",
    "demographic.vital_status",
    "demographic.days_to_death",
    "diagnoses.days_to_last_follow_up",
    "demographic.age_at_index",
    "demographic.gender",
    "demographic.race"
]].copy()

gbm_clean.columns = [
    "Patient_ID",
    "vital_status",
    "days_to_death",
    "days_to_last_follow_up",
    "age",
    "gender",
    "race"
]

# Convert numeric
gbm_clean["days_to_death"] = pd.to_numeric(gbm_clean["days_to_death"], errors="coerce")
gbm_clean["days_to_last_follow_up"] = pd.to_numeric(gbm_clean["days_to_last_follow_up"], errors="coerce")
gbm_clean["age"] = pd.to_numeric(gbm_clean["age"], errors="coerce")

# Survival time
gbm_clean["Overall_Survival_Days"] = gbm_clean["days_to_death"]
gbm_clean["Overall_Survival_Days"] = gbm_clean["Overall_Survival_Days"].fillna(
    gbm_clean["days_to_last_follow_up"]
)

# Event
gbm_clean["Event"] = gbm_clean["vital_status"].apply(
    lambda x: 1 if x == "Dead" else 0
)

# Keep required columns
gbm_clean = gbm_clean[[
    "Patient_ID",
    "Overall_Survival_Days",
    "Event",
    "age",
    "gender",
    "race"
]]

# Remove rows without survival
gbm_clean = gbm_clean.dropna(subset=["Overall_Survival_Days"])

# Drop duplicates (CRITICAL)
gbm_clean = gbm_clean.drop_duplicates(subset=["Patient_ID"])

print("GBM patients:", len(gbm_clean))
print(gbm_clean["race"].value_counts(dropna=False))
```

```python
import numpy as np

lgg = pd.read_csv("/content/drive/MyDrive/clinical.project-tcga-lgg.2026-03-03/clinical.tsv", sep="\t")

# Replace missing indicators
lgg = lgg.replace("'--", np.nan)
lgg = lgg.replace("--", np.nan)

lgg_clean = lgg[[
    "cases.submitter_id",
    "demographic.vital_status",
    "demographic.days_to_death",
    "diagnoses.days_to_last_follow_up",
    "demographic.age_at_index",
    "demographic.gender",
    "demographic.race"
]].copy()

lgg_clean.columns = [
    "Patient_ID",
    "vital_status",
    "days_to_death",
    "days_to_last_follow_up",
    "age",
    "gender",
    "race"
]

# Convert numeric
lgg_clean["days_to_death"] = pd.to_numeric(lgg_clean["days_to_death"], errors="coerce")
lgg_clean["days_to_last_follow_up"] = pd.to_numeric(lgg_clean["days_to_last_follow_up"], errors="coerce")
lgg_clean["age"] = pd.to_numeric(lgg_clean["age"], errors="coerce")

# Survival time
lgg_clean["Overall_Survival_Days"] = lgg_clean["days_to_death"]
lgg_clean["Overall_Survival_Days"] = lgg_clean["Overall_Survival_Days"].fillna(
    lgg_clean["days_to_last_follow_up"]
)

# Event
lgg_clean["Event"] = lgg_clean["vital_status"].apply(
    lambda x: 1 if x == "Dead" else 0
)

# Keep required columns
lgg_clean = lgg_clean[[
    "Patient_ID",
    "Overall_Survival_Days",
    "Event",
    "age",
    "gender",
    "race"
]]

# Remove rows without survival
lgg_clean = lgg_clean.dropna(subset=["Overall_Survival_Days"])

# Drop duplicates (CRITICAL)
lgg_clean = lgg_clean.drop_duplicates(subset=["Patient_ID"])

print("LGG patients:", len(lgg_clean))
print(lgg_clean["race"].value_counts(dropna=False))
```

```python
gbm_clean["Tumor_Type"] = 1   # GBM
lgg_clean["Tumor_Type"] = 0   # LGG
```

```python
clinical_merged = pd.concat([gbm_clean, lgg_clean], axis=0)

print("Total Patients:", len(clinical_merged))
print("Race distribution:\n", clinical_merged["race"].value_counts(dropna=False))
```

```python
clinical_merged["race"] = clinical_merged["race"].replace(
    ["not reported", "Unknown"],
    np.nan
)

# Merge small categories
clinical_merged["race"] = clinical_merged["race"].replace(
    ["american indian or alaska native"],
    "other"
)

# Fill missing
clinical_merged["race"] = clinical_merged["race"].fillna("other")
```

```python
clinical_merged["gender"] = clinical_merged["gender"].map({
    "male": 1,
    "female": 0
})
```

```python
clinical_merged = pd.get_dummies(
    clinical_merged,
    columns=["race"],
    drop_first=True
)
```

```python
print(clinical_merged.shape)
print(clinical_merged.head())
```

```python
race_cols = [
    "race_black or african american",
    "race_other",
    "race_white"
]

clinical_merged[race_cols] = clinical_merged[race_cols].astype(int)
```

```python
clinical_merged.dtypes
```

```python
final_dataset = clinical_merged.merge(
    mutation_matrix,
    on="Patient_ID",
    how="left"
)

final_dataset = final_dataset.fillna(0)

print(final_dataset.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

clinical_cols = ["age"]

final_dataset[clinical_cols] = scaler.fit_transform(
    final_dataset[clinical_cols]
)
```
